```python
from random import choice 
from itertools import combinations

from Bio import SeqIO
import subprocess
from io import StringIO

from skbio import Sequence
from skbio.sequence.distance import hamming 
#hamming (Sequence(str(A)),Sequence(str(B)))

seqs = list(SeqIO.parse('seqs_ali.fasta','fasta'))

def distancia_una(): 
    parejas = list(combinations(range(len(seqs)),2))
    pareja = choice(parejas)
    seq1 = seqs[pareja[0]]
    seq2 = seqs[pareja[1]]
    conast1 = str(seq1.seq)
    conast2 = str(seq2.seq)
    if len(conast1) == len(conast2):
        return hamming(Sequence(conast1),Sequence(conast2))


distancia_una()


mil_veces = [distancia_una() for _ in range(5000)]

import numpy as np
import matplotlib.pyplot as plt

#counts, bins = np.histogram(mil_veces, bins=40)
#plt.stairs(counts, bins)
#plt.title("Histograma")


# Plot histogram of the Hamming distances
plt.hist(mil_veces, bins=10, edgecolor='black')
plt.xlabel('Hamming Distance')
plt.ylabel('Frequency')
plt.title('Histogram of Hamming Distances')
plt.grid(True)
plt.show()
```
