```python
from random import choice 
from itertools import combinations

from Bio import SeqIO
import subprocess
from io import StringIO

from skbio import Sequence
from skbio.sequence.distance import hamming 
#hamming (Sequence(str(A)),Sequence(str(B)))

seqs = list(SeqIO.parse('seqs_ali.fasta','fasta'))

def distancia_una(): 
    parejas = list(combinations(range(len(seqs)),2))
    pareja = choice(parejas)
    seq1 = seqs[pareja[0]]
    seq2 = seqs[pareja[1]]
    conast1 = str(seq1.seq)
    conast2 = str(seq2.seq)
    if len(conast1) == len(conast2):
        return hamming(Sequence(conast1),Sequence(conast2))


distancia_una()


mil_veces = [distancia_una() for _ in range(5000)]

import numpy as np
import matplotlib.pyplot as plt

#counts, bins = np.histogram(mil_veces, bins=40)
#plt.stairs(counts, bins)
#plt.title("Histograma")


# Plot histogram of the Hamming distances
plt.hist(mil_veces, bins=20, edgecolor='black')
plt.xlabel('Hamming Distance')
plt.ylabel('Frequency')
plt.title('Histogram of Hamming Distances')
plt.grid(True)
plt.show()
```

```python
from random import choice 
from itertools import combinations
from Bio import SeqIO
from skbio import Sequence
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Hamming distance (counting differences directly)
def distancia_una():
    seqs = list(SeqIO.parse('seqs_ali.fasta', 'fasta'))  # Load sequences from FASTA file
    parejas = list(combinations(range(len(seqs)), 2))    # Generate pairs of sequence indices
    
    pareja = choice(parejas)  # Choose a random pair of sequences
    seq1 = str(seqs[pareja[0]].seq)
    seq2 = str(seqs[pareja[1]].seq)
    
    # Ensure sequences are of equal length
    if len(seq1) != len(seq2):
        return None
    
    # Initialize a counter for differences
    num_differences = 0
    
    # Compare characters at each position
    for i in range(len(seq1)):
        if seq1[i] != seq2[i]:
            num_differences += 1
    
    return num_differences

# Generate 5000 Hamming distances
veces = [distancia_una() for _ in range(10000)]

# Plot histogram of the Hamming distances
plt.hist(veces, bins=20, edgecolor='black')
plt.xlabel('Hamming Distance')
plt.ylabel('Frequency')
plt.title('Histogram of Hamming Distances')
plt.grid(True)
plt.show()
```
