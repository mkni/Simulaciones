```python
#Nuevo codigo con los mismos datos de arriba

from random import choice 
from itertools import combinations
from Bio import SeqIO
from skbio import Sequence
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Hamming distance (counting differences directly)
def distancia_una():
    seqs = list(SeqIO.parse('seqs_ali.fasta', 'fasta'))  # Load sequences from FASTA file
    parejas = list(combinations(range(len(seqs)), 2))    # Generate pairs of sequence indices
    
    pareja = choice(parejas)  # Choose a random pair of sequences
    seq1 = str(seqs[pareja[0]].seq)
    seq2 = str(seqs[pareja[1]].seq)
    
    # Ensure sequences are of equal length
    if len(seq1) != len(seq2):
        return None
    
    # Initialize a counter for differences
    num_differences = 0
    
    # Compare characters at each position
    for i in range(len(seq1)):
        if seq1[i] != seq2[i]:
            num_differences += 1
    
    return num_differences

# Generate 5000 Hamming distances
veces = [distancia_una() for _ in range(10000)]

# Plot histogram of the Hamming distances
plt.hist(veces, bins=20, edgecolor='black')
plt.xlabel('Hamming Distance')
plt.ylabel('Frequency')
plt.title('Histogram of Hamming Distances')
plt.grid(True)
plt.show()
```
