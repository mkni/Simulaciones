```python
def distancia_una(): 
    parejas = list(combinations(range(len(seqs)),2))
    pareja = choice(parejas)
    seq1 = seqs[pareja[0]]
    seq2 = seqs[pareja[1]]
    conast1 = str(seq1.seq)
    conast2 = str(seq2.seq)
    if len(conast1) == len(conast2):
        return hamming(Sequence(conast1),Sequence(conast2))
```

```python
distancia_una()
```

```python
mil_veces = [distancia_una() for _ in range(100000)]
```
